```python
class UPI:
  def pay(self):
    print("Payment through UPI")
class Card:
  def pay(self):
    print("Payment through Card")
for payment in [UPI(),Card()]:
  payment.pay()
```

```python
expenses = {}

def add_expense(category, amount):
    if category in expenses:
        expenses[category] += amount
    else:
        expenses[category] = amount
    print(f"Added {amount:.2f} to {category}.")

def view_expenses():
    if not expenses:
        print("No expenses recorded yet.")
        return
    print("\n--- Current Expenses ---")
    for category, amount in expenses.items():
        print(f"{category}: ${amount:.2f}")
    print("------------------------")

def summarize_expenses():
    total = sum(expenses.values())
    print(f"\nTotal expenses: ${total:.2f}")

# Demo usage:
add_expense('Groceries', 50.75)
add_expense('Transport', 15.20)
add_expense('Groceries', 30.00)
add_expense('Utilities', 75.50)

view_expenses()
summarize_expenses()

```
